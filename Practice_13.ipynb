```python
# df["total_bill"]
# df["tip"]

mean_bill=df["total_bill"].mean()
std_bill=df["total_bill"].std()
lower_fence=mean_bill-(3*std_bill)
upper_fence=mean_bill+(3*std_bill)
df_total_bill = df.loc[(df["total_bill"] > lower_fence) & (df["total_bill"] < upper_fence), "total_bill"]

df_total_bill.skew()
```
