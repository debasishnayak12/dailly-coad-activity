```python
city_start_with_vowels=df[df["City"].str.match('^[AEIOU]')]['City'].unique()
city_start_with_vowels=sorted(city_start_with_vowels)
city_start_with_vowels
```

```python
## OR 
def start_with_v(s):
    vowels='AEIOU'
    return s[0] in vowels
```

```python
strt_with_vowels=df[df["City"].apply(start_with_v)]["City"].unique()
strt_with_vowels=sorted(strt_with_vowels)
strt_with_vowels
```

```python
# City not start with vowels

def not_start_v(s):
    vowels='AEIOU'
    return s[0] not in vowels
```

```python
not_st_with_v=df[df["City"].apply(not_start_v)]["City"].unique()
not_st_with_v=sorted(not_st_with_v)
not_st_with_v
```
