```python
def live_object_detection():
    cap=cv2.VideoCapture(0)
    
    model=YOLO("yolov8n.pt")
    
    box_annatator=sv.BoxAnnotator()
    
    while True:
        ret, frame=cap.read()
        
        data=model(frame)
        
        detection=sv.Detections.from_ultralytics(data[0])
        
        if not ret:
            print("Error: Failed to grab frame  ")
            break
        frame = box_annatator.annotate(scene=frame,detections=detection)
        
        cv2.imshow("Live object detection",frame)
        
        if cv2.waitKey(1) & 0xFF==ord("c"):
            break
        
    cap.release()
    cv2.destroyAllWindows()
    
live_object_detection()
     
```

```python
def live_object_detection():
    from ultralytics import YOLO
    import cv2
    import numpy as np

    # Load the YOLO model (YOLOv8 in this case)
    model = YOLO('yolov8n.pt') 

    # Open external webcam (change the index if necessary)
    cap = cv2.VideoCapture(1)  # Use the correct index for your external camera

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Apply Gaussian Blur to reduce noise
        blurred_frame = cv2.GaussianBlur(frame, (5, 5), 0)

        # Convert the frame to HSV color space
        hsv_frame = cv2.cvtColor(blurred_frame, cv2.COLOR_BGR2HSV)

        # Define range for white color and create a mask
        lower_white = np.array([0, 0, 168])  # Adjusted lower bound
        upper_white = np.array([172, 111, 255])  # Adjusted upper bound
        white_mask = cv2.inRange(hsv_frame, lower_white, upper_white)

        # Apply morphological operations to clean up the mask
        # kernel = np.ones((5, 5), np.uint8)
        # white_mask = cv2.erode(white_mask, kernel, iterations=2)
        # white_mask = cv2.dilate(white_mask, kernel, iterations=2)

        # Find contours in the mask
        contours, _ = cv2.findContours(white_mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

        # Initialize tracking variables
        ball_detected = False
        ball_bbox = None

        for contour in contours:
            # Calculate the area of the contour
            area = cv2.contourArea(contour)
            # Filter out small contours (if necessary)
            if area > 500:
                # Get the bounding box coordinates
                x, y, w, h = cv2.boundingRect(contour)
                ball_bbox = (x, y, w, h)
                ball_detected = True
                # Draw a blue rectangle around the white ball with thicker border
                cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 3)
                # Draw the contour with red color and thinner lines
                #cv2.drawContours(frame, [contour], -1, (0, 0, 255), 1)

        # if not ball_detected:
        #     # If no ball detected, use the previous bounding box (if available)
        #     if ball_bbox is not None:
        #         x, y, w, h = ball_bbox
        #         cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
        
        # Display the frame
        cv2.imshow("Real-time Object Detection", frame)

        if cv2.waitKey(1) & 0xFF == ord('c'):
            break

    cap.release()
    cv2.destroyAllWindows()
    
live_object_detection()
```
