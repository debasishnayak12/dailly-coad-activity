```python
import logging 
from bs4 import BeautifulSoup
import os
import requests
```

```python
save_dir="images"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
    
```

```python
query="Virat Kohli"
url="https://www.google.com/search?sxsrf=AB5stBgyhl4VB7NQG5BwXo3OtmsqLQVIPw:1691496369907&q=virat+kohli&tbm=isch&source=lnms&sa=X&ved=2ahUKEwiShJGCg82AAxW43TgGHcOJB34Q0pQJegQIDRAB&biw=1280&bih=595&dpr=1.5"
```

```python
response=requests.get(f"https://www.google.com/search?sxsrf=AB5stBgyhl4VB7NQG5BwXo3OtmsqLQVIPw:1691496369907&q={query}&tbm=isch&source=lnms&sa=X&ved=2ahUKEwiShJGCg82AAxW43TgGHcOJB34Q0pQJegQIDRAB&biw=1280&bih=595&dpr=1.5")
response
```

```python
soup=BeautifulSoup(response.content,"html.parser")
soup
```

```python
image_tag=soup.find_all("img")[2:]
image_tag
```

```python
for i in image_tag:
    image_url=i["src"]
    image_data=requests.get(image_url).content
    with open(os.path.join(save_dir,f"{query}_{image_tag.index(i)}.jpg"),"wb")as f:
        f.write(image_data)
```
